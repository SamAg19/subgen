```python
import ffmpeg
import pandas as pd
import numpy as np
import os
import librosa
```

```python
x, sr = librosa.load('Interstellar (2014) (2014) [1080p]/output-audio.wav')
```

```python
sr
```

```python
mfccs = librosa.feature.mfcc(x, sr=sr)
print(mfccs.shape)
```

```python
import sklearn
```

```python
mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
```

```python
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```

```python
mfccs1 = librosa.feature.mfcc(x[int(75.5*sr):int(76.5*sr)], sr=sr)
```

```python
mfccs1 = sklearn.preprocessing.scale(mfccs1, axis=1)
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs1, x_axis='time',)
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
